```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
X = df.iloc[:, [3, 4]].values
X
```

```python
plt.scatter(X[:,0], X[:,1])
plt.title('Income VS Spending')
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5) 
```

```python
y_pred = km.fit_predict(X)
y_pred
```

```python
# combine x and ypred in one array
x = np.column_stack((X, y_pred))
x
```

```python
# find the centroids of cluster
km.cluster_centers_
```

```python
# to separate the clusters
x1 = x[x[:,2]==0]
x2 = x[x[:,2]==1]
x3 = x[x[:,2]==2]
x4 = x[x[:,2]==3]
x5 = x[x[:,2]==4]

plt.scatter(x1[:,0], x1[:,1], color='red', label = 'Cluster 1')
plt.scatter(x2[:,0], x2[:,1], color='green', label = 'Cluster 2')
plt.scatter(x3[:,0], x3[:,1], color='yellow', label = 'Cluster 3')
plt.scatter(x4[:,0], x4[:,1], color='pink', label = 'Cluster 4')
plt.scatter(x5[:,0], x5[:,1], color='blue', label = 'Cluster 5')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s=100, color='purple', marker='*', label='Centroids')

plt.xlabel('annual Income')
plt.ylabel('Spending score')
plt.legend()
```

```python
sse = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', random_state=42)
    km.fit(X)
    sse.append(km.inertia_) 
    # function inertia_ is sum of square of errors
sse
```

```python
plt.plot(range(1, 11), sse)
plt.title('Elbow method')
plt.xlabel('k cluster')
plt.ylabel('sse')
```
