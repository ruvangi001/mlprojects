```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
df.head()
```

```python
l = []
for i in range(0,7501):
    l.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions=l, min_support = 0.003, min_confidence=0.2, min_lift=3, min_length=2,max_length=2)
```

```python
results = list(rules)
print(results)
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]    
    support = [result[1] for result in results]    
    confidence = [result[2][0][2] for result in results] 
    lift = [result[2][0][3] for result in results]  
    return list(zip(lhs, rhs, support, confidence, lift))
rsdf = pd.DataFrame(inspect(results), columns=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
rsdf
```

```python
rsdf.nlargest(n=10, columns='Lift')
```
