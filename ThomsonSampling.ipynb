```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ads_CTR_Optimisation.csv')
df
```

```python
import random
N = 1000
d = 10
ads_selected = []
noofrewards_1 = [0] * d
noofrewards_0 = [0] * d
totalreward = 0

for n in range(0, N):
    ad = 0
    max_random = 0
    # print('reward 1:- ',noofrewards_1)
    # print('reward 0:- ',noofrewards_0)

    for i in range(0, d):
        randombeta = random.betavariate(noofrewards_1[i] + 1, noofrewards_0[i] + 1)
        # print('randombeta:- ',randombeta)
        if randombeta > max_random:
            max_random = randombeta
            ad = i
            # print('max_random:- ',max_random)
            # print('ad:- ',ad)
    ads_selected.append(ad)
    # print('ads_selected:- ', ads_selected)
    reward = df.values[n, ad]
    # print('reward:- ',reward)
    if reward == 1:
        noofrewards_1[ad] = noofrewards_1[ad] + 1
    else:
        noofrewards_0[ad] = noofrewards_0[ad] + 1
    totalreward = totalreward + reward
    # print('totalreward:- ',totalreward)
    # print('\n')
    
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selection')
plt.xlabel('Ads')
plt.ylabel('No of times ads selected')
plt.show()
```
